```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling parameters
fs = 1000              # samples per second
duration = 5.0        # seconds
t = np.linspace(0, duration, int(fs*duration))

# Frequency parameters
f0 = 1                 # start frequency (Hz)
f1 = 20                # end frequency (Hz)

# Linear frequency variation rate
k = (f1 - f0) / duration

# Phase computation (integral of frequency)
phase = 2*np.pi*(f0*t + 0.5*k*t**2)

# Signal
x = np.sin(phase)

# Plot signal
plt.figure(figsize=(12,4))
plt.plot(t, x)
plt.title("Signal with Increasing Frequency (Chirp)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
